SLICING OUT USEFUL COLUMNS FROM DATASET 

```python
# CREATING AN ARRAY CONTAING LIST OF AGE AND BP EACH
```

```python
# iloc[row,column]->locating specific columns
# all rows of column 0 and column 3
x=dataset.iloc[:,[0,3]].values
print('        AGE    BP')
x
```

```python
# finding wcss value [elbow method]
# wcss=within clusters sum of squared distance
```

```python

wcss=[]

for i in range(1,16):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=50)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
print(wcss)
```

```python
# PLOTTING THE ELBOW GRAPH
```

```python
sns.set()  #gives basic thems and basic parameters for required for the graph
plt.plot(range(1,16),wcss)
plt.title('The Elbow Point Graph',color="green",size=30)
plt.xlabel('Number of Clusters',color="blue",size=20)
plt.ylabel('WCSS',color="blue",size=20)
plt.show()
```

```python
# NUMBER OF CLUSTERS:6
# AFTER 6 not much difference
```

```python
# loading our model to the variable kmeans
kmeans=KMeans(n_clusters=6,init='k-means++',random_state=0)

# returning a label for each data point based on their 
# cluster->all 6 clusters will be given a label
# fit_predict ->return cluster number for each data point
y=kmeans.fit_predict(x)
print("Y:\n",y)
```

```python
# plotting all the clusters and their centroids
plt.figure(figsize=(8,8))#->figure size
# X:0,Y:1        x             y
plt.scatter(x[y == 0, 0], x[y == 0, 1], color='green', label='Cluster 1')
plt.scatter(x[y == 1, 0], x[y == 1, 1], color='brown', label='Cluster 2')
plt.scatter(x[y == 2, 0], x[y == 2, 1], color='yellow', label='Cluster 3')
plt.scatter(x[y == 3, 0], x[y == 3, 1], color='blue', label='Cluster 4')
plt.scatter(x[y == 4, 0], x[y == 4, 1], color='orange', label='Cluster 5')
plt.scatter(x[y == 5, 0], x[y == 5, 1], color='violet', label='Cluster 6')


# plotting centroids,s=size
#                                  x axis:0                     y axis:1        
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=100,color='black',label='Centroids')
plt.title('BP v/s AGE ',size=30,color='green')
plt.xlabel('AGE',color="blue",size=20)
plt.ylabel("BP",color="blue",size=20)
plt.legend()
plt.show()
```
